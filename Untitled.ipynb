```python
# Data PreProcessing
# No missing value to dealing with
# Dealing with Outliers
def outlier_thresholds(dataframe, col_name, q1=0.25, q3=0.75):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit
     
    
def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False

    
def grab_outliers(dataframe, col_name, index=False):
    low, up = outlier_thresholds(dataframe, col_name)

    if dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))].shape[0] > 10:
        print(dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))].head())
    else:
        print(dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))])

    if index:
        outlier_index = dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))].index
        return outlier_index
    
def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
    
    
for col in num_cols:
    replace_with_thresholds(df, col)
```

```python
# Get a training and test dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df,
                                                    y,
                                                    stratify=y,
                                                    random_state = 152,
                                                    test_size = 0.2, 
                                                    shuffle = True)

print(f"The shape of x_train is {x_train.shape}")
print(f"The shape of x_test is {x_test.shape}")
print(f"The shape of y_train is {y_train.shape}")
print(f"The shape of y_test is {y_test.shape}")
```

```python
from sklearn.metrics import f1_score,accuracy_score,roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Perceptron
from sklearn.neighbors import *
import plotly.express as px
import plotly.offline as py
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from plotly.offline import init_notebook_mode, iplot

def predictions(model):
    
    y_pred = model.fit(x_train, y_train).predict(x_test)
    Accuracy = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, model.predict_proba(x_test)[:,1])
    f1 = f1_score(y_test, y_pred)
    scores = pd.DataFrame({'Scores': [Accuracy, roc_auc, f1],
                            'Metrics': ['Accuracy',
                                        'ROC-AUC',
                                        'F1-Score']})
    
    fig = make_subplots(rows = 1, cols = 1)
    fig.add_trace(go.Bar(x = [round(i,5) for i in scores['Scores']],
                         y = scores['Metrics'],
                         text = [round(i,5) for i in scores['Scores']],
                         orientation='h',
                         textposition = 'inside',
                         name = 'Scores',
                         marker = dict(color = metric_constraints,
                                       line_color = 'white',
                                       line_width=1.5)),
                  row = 1, col = 1)
    fig.update_layout(title={'text': model.__class__.__name__ ,
                             'y':0.9,
                             'x':0.5,
                             'xanchor': 'center',
                             'yanchor': 'top'},
                      template='plotly_white')
    fig.update_xaxes(range=[0,1], row = 1, col = 1)

    iplot(fig)

_models = [LogisticRegression(),
           KNeighborsClassifier(n_neighbors= 3)
            ]

for model in _models:
    predictions(model)
```

```python
# DISCUSSION OF QUESTION 2: 

# Logistic Regression is a linear classifier nd has ability to model the patterns of the indepeendent features and 
# probability of their belonging class. Therefore, it catches the linear relationships and models them. 
# However, for catching the non-linear relations and patterns other algorithms may  be better like k-NN. 

# k-NN (k-Nearest Neighbors) is a non-parametric supervised learning algorithm.
# It classifies data points based on similarity to other points. 
# It assumes that similar data points are near the each other. 
# It shows the local patterns / characteristics of the dataset. 
# When making predictions for a given observation, the k-NN algorithm identifies K different observation units in the dataset 
# that are closest to the target observation. 
# These K nearest neighbors are selected based on their feature values. 
# Once the K nearest neighbors are identified, the algorithm uses their corresponding dependent variable values to
# make predictions for the target observation.
# It can be costly depending on dataset and number of the features. 


# The Perceptron algorithm is a simple binary classification machine learning algorithm. It is a linear classifier. 
# It can cope with linearly separable datasets, but other algorithms can be better for non-linear relations. 
# It is the simplest (basic) neural network that single neuron that takes a row of data as input and predicts a class label

# Logistic Regression tends to perform well when the classes are well-separated, 
# k-NN can be effective when there are local patterns, and the Perceptron algorithm can work well for linearly 
# separable datasets.

# IN THIS CASE, ACCURACY VALUES OF THE THREE ALGORITHM IS VERY SIMILAR. HOWEVER IN CYBER SECURITY RELATED PROBLEMS 
# USING RECALL VALUE RATHER THAN ACCURACY CAN BE MORE ACCURATE. HOWEVER, WE GOT VERY SIMILAR THREE ACCURACY VALUES BECAUSE 
# DATASET CONTAINS LINEAR RELATIONS AND IT IS LINEARLY SEPARABLE. DATASET IS NOT VERY COMPLICATED AND PROBLEM IS SIMPLE. 
# CLASSES ARE WELL-SEPARABLE. THEIR DECISION BOUNDRIES TO CLASSIFY PROBLEM MAY SIMILAR. WE PROCESSED THE DATASET BEFORE 
# MAKING PREDICTIONS. IT MAY AFFECT QUALITY OF PREDICTIONS IN A POSITIVE WAY.
```
